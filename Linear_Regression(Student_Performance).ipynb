```python
import numpy as np
import pandas as pd
df=pd.read_csv('/content/Student_Performance.csv')
df
```

```python
from sklearn.preprocessing import LabelEncoder
lab=LabelEncoder()
df['Extracurricular Activities']=lab.fit_transform(df['Extracurricular Activities'])
df['Extracurricular Activities']
```

```python
df.dtypes
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
y_test
```

```python
print(model.predict([[5,87,0,5,3]]))
```

```python
print('Slope is')
list(zip(x,model.coef_))
```

```python
print('Constant is',model.intercept_)
```

```python
df1=pd.DataFrame({'Actual_value':y_test,'Predicted_value':y_pred,'Difference':y_test-y_pred})
df1
```

```python
from sklearn.metrics import mean_absolute_error,mean_absolute_percentage_error,mean_squared_error,r2_score
print('Error           :',mean_absolute_error(y_test,y_pred))
print('Percentage Error:',mean_absolute_percentage_error(y_test,y_pred))
print('Squared Error   :',mean_squared_error(y_test,y_pred))
print('R2_score        :',r2_score(y_test,y_pred)*100)
```
