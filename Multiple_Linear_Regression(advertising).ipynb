```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

Model Creation

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
print(model.predict([[150,55,72]]))
```

```python
print('Slope is',model.coef_) #here we got coefficients of all independent variables but need to specify those according w.r.to resepctive
# variable names
```

```python
print('Slope is')
list(zip(x,model.coef_))
```

```python
print('Constant is',model.intercept_)
```

```python
df1=pd.DataFrame({'Actual_value':y_test,'Predicted_value':y_pred,'Difference':y_test-y_pred})
df1
```

Performance Evaluation

```python
# Performance Evaluation
from sklearn.metrics import mean_absolute_error
print('Error is',mean_absolute_error(y_test,y_pred))
```

```python
from sklearn.metrics import mean_absolute_percentage_error
print('Percentage Error',mean_absolute_percentage_error(y_test,y_pred))
```

```python
from sklearn.metrics import mean_squared_error
print('Squared Error',mean_squared_error(y_test,y_pred))
```

```python
data=mean_squared_error(y_test,y_pred)
data
```

```python
sq=np.sqrt(data)
sq
```

```python
from sklearn.metrics import r2_score
print('R2_score',r2_score(y_test,y_pred)*100)
```
